```python
import pandas as pd

tabela = pd.read_csv('exportacoes_franca.csv')
display(tabela)
```

### Como foi a evolução das exportações para a frança ao longo dos anos?

```python
tabela_anos_expo = tabela[['Year', 'US$ FOB']].groupby("Year").sum()

def format_valor(valor):
    valor_formatado = "US$ {:,.2f}".format(valor)
    return valor_formatado

tabela_anos_expo['US$ FOB'] = tabela_anos_expo['US$ FOB'].apply(format_valor)
display((tabela_anos_expo))
```

### Quais os produtos mais exportados ao longo de todo o período?

```python
tabela_produtos = tabela[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum()
tabela_produtos= tabela_produtos.sort_values(by='US$ FOB',ascending=False)

tabela_produtos['US$ FOB'] = tabela_produtos['US$ FOB'].apply(format_valor)

display(tabela_produtos[:])
```

### Em 2020 qual cidade mais exportou para a França?
Em valores US$

```python
tabela_2020 = tabela.loc[tabela['Year']==2020, :]

tabela_2020 = tabela_2020[['City', 'US$ FOB']].groupby('City').sum()
tabela_2020 = tabela_2020.sort_values(by='US$ FOB', ascending=False)

tabela_2020['US$ FOB'] = tabela_2020['US$ FOB'].apply(format_valor)
display(tabela_2020)
```

### Quais os produtos mais exportados (em US$) que as 2 maiores cidades (em exportação em 2020) exportaram?

```python
tabela_duque = tabela.loc[tabela['Year']==2020]
tabela_duque = tabela_duque.loc[tabela_duque['City'] =='Duque de Caxias - RJ']
tabela_duque = tabela_duque[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum(numeric_only=True)
tabela_duque = tabela_duque.sort_values(by='US$ FOB', ascending=False)

tabela_lem = tabela.loc[tabela['Year']==2020]
tabela_lem = tabela_lem.loc[tabela_lem['City']=='Luís Eduardo Magalhães - BA']
tabela_lem = tabela_lem[['SH2 Description', 'US$ FOB']].groupby('SH2 Description').sum(numeric_only=True)
tabela_lem = tabela_lem.sort_values(by='US$ FOB', ascending=False)

tabela_duque['US$ FOB'] = tabela_duque['US$ FOB'].apply(format_valor)
tabela_lem['US$ FOB'] = tabela_lem['US$ FOB'].apply(format_valor)
print("\t\t\t\tDUQUE DE CAXIAS (RJ)")
display(tabela_duque)
print("\t\t\t\t LUÍS EDUARDO MAGALHÃES (BA)")
display(tabela_lem)
```
